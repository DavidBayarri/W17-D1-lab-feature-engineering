```python
features_to_drop = null_values[null_values['%'] > 80]
features_to_drop_list = list(features_to_drop.index)
features_to_drop_list
```

```python
features_to_analyze = null_values[null_values['%'].between(1,80)]
features_to_analyze_list = list(features_to_analyze.index)
features_to_analyze_list
```
